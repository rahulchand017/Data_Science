## 1. Relational Plot
- To see the statistical relation between 2 or more variables.
- Bivariate Analysis

#### Plots under this sectin
- Scatter plot
- Lineplot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Scatter plot
tips = pd.read_csv('tips.csv')
tips
```

```python
tips.info()
```

```python
# scatter plot -> axes level function
sns.scatterplot(data = tips, x='total_bill', y = 'tip', hue ='sex' )
```

```python
# Scatter using relplot -> replot is a figure level function
sns.relplot(data= tips, x='total_bill', y='tip', kind = 'scatter', hue ='sex', style='time')
```

```python
# Line plot
gap = pd.read_csv('gapminder.csv')
temp_df = gap[gap['country']== 'India']
temp_df
```

```python
sns.lineplot(data = temp_df, x='year', y='lifeExp')
```

```python
# using relplot
sns.relplot(data =temp_df, x='year', y='lifeExp', kind = 'line')
```

```python
temp_df = gap[gap['country'].isin(['India','Pakistan','China'])]
```

```python
sns.relplot(kind='line', hue='country', x='year',data=temp_df, y='lifeExp')
```

```python
# facat plot work only with figure level function
sns.relplot(data=tips, kind='scatter',x='total_bill', y ='tip', col='sex', hue ='sex', row='smoker')
```

# Distribution plot


```python
# figure level function-> distplot
# axis level function -> histplot -> Kdeplot -> rugplot
```

```python
# axis level plot
sns.histplot(data = tips, x ='total_bill')
```

```python
# figure level function
sns.displot(data=tips, x='total_bill', kind ='hist')
```
