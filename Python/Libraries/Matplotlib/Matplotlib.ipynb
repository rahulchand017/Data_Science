```python
# multiple bar chart
import matplotlib.pyplot as plt
import numpy as np

# Extract data from the DataFrame
batsmen = df['batsman']
seasons = ['2015', '2016', '2017']
x = np.arange(len(batsmen))  # Position of bars on x-axis
width = 0.25  # Width of each bar

# Create the multiple bar chart
plt.bar(x - width, df['2015'], width, label='2015')
plt.bar(x, df['2016'], width, label='2016')
plt.bar(x + width, df['2017'], width, label='2017')

# Add labels, title, and legend
plt.xticks(x, batsmen, rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.xlabel('Batsman')
plt.ylabel('Runs')
plt.title('Performance of Batsmen Across Seasons')
plt.legend(title='Season')
plt.tight_layout()  # Adjust layout to prevent clipping

# Display the plot
plt.show()
```

```python
from matplotlib import cm

plt.style.use('_mpl-gallery')

# Make data
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z, vmin=Z.min() * 2, cmap=cm.Blues)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```

### 3D Surface plot

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y)
yy.shape
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig=plt.figure(figsize=(12,8))
ax = plt.subplot(projection = '3d')
ax.plot_surface(xx,yy,z, cmap='coolwarm')
```

### Contour Plot
- Represent 3D graph in 2D

```python
fir = plt.figure(figsize = (12,8))
ax = plt.subplot()
p = ax.contour(xx,yy,z)
fig.colorbar(p, cmap = 'coolwarm')
plt.title("Contour graph")
```

```python
fig = plt.figure(figsize=(12, 8))  # Use 'fig' consistently
ax = fig.add_subplot(111)  # Explicitly add the axis to the correct figure
p = ax.contourf(xx, yy, z, cmap='coolwarm')
fig.colorbar(p, ax=ax)  # Attach the colorbar to the correct figure and axis
plt.title("ContourF graph")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

import matplotlib
import matplotlib as mpl


vegetables = ["cucumber", "tomato", "lettuce", "asparagus",
              "potato", "wheat", "barley"]
farmers = ["Farmer Joe", "Upland Bros.", "Smith Gardening",
           "Agrifun", "Organiculture", "BioGoods Ltd.", "Cornylee Corp."]

harvest = np.array([[0.8, 2.4, 2.5, 3.9, 0.0, 4.0, 0.0],
                    [2.4, 0.0, 4.0, 1.0, 2.7, 0.0, 0.0],
                    [1.1, 2.4, 0.8, 4.3, 1.9, 4.4, 0.0],
                    [0.6, 0.0, 0.3, 0.0, 3.1, 0.0, 0.0],
                    [0.7, 1.7, 0.6, 2.6, 2.2, 6.2, 0.0],
                    [1.3, 1.2, 0.0, 0.0, 0.0, 3.2, 5.1],
                    [0.1, 2.0, 0.0, 1.4, 0.0, 1.9, 6.3]])


fig, ax = plt.subplots()
im = ax.imshow(harvest)

# Show all ticks and label them with the respective list entries
ax.set_xticks(range(len(farmers)), labels=farmers,
              rotation=45, ha="right", rotation_mode="anchor")
ax.set_yticks(range(len(vegetables)), labels=vegetables)

# Loop over data dimensions and create text annotations.
for i in range(len(vegetables)):
    for j in range(len(farmers)):
        text = ax.text(j, i, harvest[i, j],
                       ha="center", va="center", color="w")

ax.set_title("Harvest of local farmers (in tons/year)")
fig.tight_layout()
plt.show()
```
