The Parametric Density Estimation, is a method used to estimate the probability density fnx of a R.V by assuming that the underlying distribution belongs to a specific parametric family of probability distributions, such as normal, exponential, poisson distribution.

### Gaussian Mixture Models(GMMs) another name of Normal Distribution

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import normal

sample = normal(loc=50, scale=5, size=1000)
```

```python
sample.mean()
```

```python
# plot histogram to understand the distribution of data
plt.hist(sample, bins=10)
# plot is approx distribution
```

```python
sample_mean= sample.mean()
sample_sd = sample.std()
```

 Normal distribution have parameter mean, S.D

```python
from scipy.stats import norm
dist = norm(sample_mean, sample_sd)
```

```python
values = np.linspace(sample.min(), sample.max(),100)
```

```python
prob_density = [dist.pdf(value) for value in values]
```

```python
# Plot the histogram and pdf
plt.hist(sample, bins=10, density = True)
plt.plot(values, prob_density)
```
