A non-parametric density estimate is a way to determine the probability distribution of data by constructing the probability density function directly from the data itself, without assuming a specific mathematical form (like a normal or exponential distribution).

#### Popular methods include histograms, which divide data into bins, and kernel density estimation (KDE), where a smoothing function (kernel) is centered on each data point, with the total density being the sum of these kernels.

## Kernel Density Estimator (KDE)
- The KDE technique involves using a kernel fxn to smooth out the data and create a continuous estimate of the underlying density fxn.
