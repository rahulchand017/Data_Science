```python
# Changing line style

plt.plot(batsman['index'], batsman['V Kohli'], color = 'Purple', linestyle = 'dashed') #dashdot
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'Red', linestyle = 'dotted')
plt.title("Comparisons between Virat and Rohit")
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# Changing marker(points)

plt.plot(batsman['index'], batsman['V Kohli'], color = 'Purple', linestyle = 'dashed', marker = 'D') #dashdot #markersize=1
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'Red', linestyle = 'dotted', marker = '>')
plt.title("Comparisons between Virat and Rohit")
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# Labeling the lines

plt.plot(batsman['index'], batsman['V Kohli'], color = 'Purple', linestyle = 'dashed', marker = 'D', label='Virat') #dashdot #markersize=1
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'Red', linestyle = 'dotted', marker = '>', label='Rohit')
plt.title("Comparisons between Virat and Rohit")
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend() # show which line is for which column
plt.grid() # Providing the grid
plt.show() # Representation of graph
```

#### Multiple plots at one's

```python
fig, ax = plt.subplots(nrows = 2, ncols=1, sharex= True)
ax[0].scatter(batters['avg'], batters['strike_rate'])
ax[0].set_title("Avg vs Strike rate")
ax[0].set_ylabel('Strike rate')


ax[1].scatter(batters['avg'], batters['runs'])
ax[1].set_title("Avg vs runs")
ax[1].set_ylabel('Runs')
ax[1]. set_xlabel('Avg')
```

## 3D Surface Plots

```python
# Loss function
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx , yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(15,7))
ax = plt.subplot(projection = '3d')
p = ax.plot_surface(xx, yy, z, cmap = 'summer')
fig.colorbar(p)
# to plt a figure bar we have assign the values to p
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

## Contour plot
- Represent the 3d into 2d plot

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
xx , yy = np.meshgrid(x,y)
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(15,7))
ax = plt.subplot()
p= ax.contour(xx, yy, z , cmap ='summer')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(15,7))
ax = plt.subplot()
p= ax.contourf(xx, yy, z , cmap ='summer') # contourf will do the filling
fig.colorbar(p)
```
