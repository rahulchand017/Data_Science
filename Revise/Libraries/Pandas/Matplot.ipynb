### 2D- Line plot
- Bivariate Analysis
- Categorical -> numerical and numerical -> numerical
- Use case - Time series data

```python
# Plotting a simple function
price = [46000, 49000, 55000, 48000, 45000, 40000]
year = [2018,2019,2020,2021,2022,2023]
plt.plot(year, price)
# Categorical to y-axis and numerical to x-axis
```

```python
batsman = pd.read_csv('C:/Users/rahul/Python_Programming/Libraries/Matplotlib/sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'], batsman['V Kohli'])
```

```python
# Plotting multiple plots
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])
```

```python
# labels title
# Color changing in the graphs

plt.plot(batsman['index'], batsman['V Kohli'], color = 'Purple')
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'Red')
plt.title("Comparisons between Virat and Rohit")
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# Changing line style

plt.plot(batsman['index'], batsman['V Kohli'], color = 'Purple', linestyle = 'dashed') #dashdot
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'Red', linestyle = 'dotted')
plt.title("Comparisons between Virat and Rohit")
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# Changing marker(points)

plt.plot(batsman['index'], batsman['V Kohli'], color = 'Purple', linestyle = 'dashed', marker = 'D') #dashdot #markersize=1
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'Red', linestyle = 'dotted', marker = '>')
plt.title("Comparisons between Virat and Rohit")
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# Labeling the lines

plt.plot(batsman['index'], batsman['V Kohli'], color = 'Purple', linestyle = 'dashed', marker = 'D', label='Virat') #dashdot #markersize=1
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'Red', linestyle = 'dotted', marker = '>', label='Rohit')
plt.title("Comparisons between Virat and Rohit")
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend() # show which line is for which column
plt.grid() # Providing the grid
plt.show() # Representation of graph
```

```python
# providing a outlier in the graph
price = [46000, 49000, 55000, 48000, 45000, 40000, 3000000]
year = [2018,2019,2020,2021,2022,2023, 2024]
plt.plot(year, price)
plt.ylim(0,60000) # trimming the values from the values
```

## Bar chart
- Bivariate, Univariate analysis
- Numerical vs Categorical
- Use case- Aggregate analysis of groups

```python
# simple bar chart
children = [5,2,6,3,1,7]
color = ['Red', 'Blue','Green', 'Pink', 'Violet', 'Black']
plt.bar(color, children)
plt.xlabel('Fav color')
plt.ylabel('Children')
```

```python
# bar using 
plt.barh(color, children)
```

```python
df = pd.read_csv('C:\\Users\\rahul\\Python_Programming\\Libraries\\Matplotlib\\batsman_season_record.csv')
df
```

```python
# multiple bar chart
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2,color='yellow')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2,color='blue')

plt.xticks(np.arange(df.shape[0]), df['batsman'])

plt.show()
```

```python
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children,color='black')
plt.xticks(rotation='vertical')

# clarity in bar is not clear so use xticks
plt.show()
```

### Stacked bar chart

```python
# Stacked Bar chart
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```

## Pie chart
- Univariate/ Bivariate analysis
- Categorical vs numerical
- Use case- To find contribution on a standard scale

```python
# simple data
data = [23,45,100, 69, 22]
subjects = ['Hindi', 'Physics', 'Maths', 'English', 'Chemistry']
plt.pie(data, labels= subjects )
plt.show()
```

```python
# dataset
df = pd.read_csv('C:\\Users\\rahul\\Python_Programming\\Libraries\\Matplotlib\\gayle-175.csv')
```
